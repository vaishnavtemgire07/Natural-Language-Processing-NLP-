```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download required resources (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Sample text
text = """Text preprocessing techniques are essential for improving
the performance of Natural Language Processing models."""

# 1. Lowercasing (Normalization)
text = text.lower()

# 2. Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# 3. Tokenization
tokens = word_tokenize(text)

# 4. Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

# 5. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

# 6. Stemming
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in lemmatized_tokens]

# Output
print("Original Text:\n", text)
print("\nTokens:\n", tokens)
print("\nAfter Stopword Removal:\n", filtered_tokens)
print("\nAfter Lemmatization:\n", lemmatized_tokens)
print("\nAfter Stemming:\n", stemmed_tokens)
```
