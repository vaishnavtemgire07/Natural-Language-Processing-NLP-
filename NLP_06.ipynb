```python
import nltk
from nltk.corpus import wordnet as wordnet
from nltk.tokenize import word_tokenize
```

```python
# download resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python
# custom lesk function
def lesk_algorithm(sentence, word):

    context = set(word_tokenize(sentence.lower()))
    synsets = wordnet.synsets(word)

    best_synset = None
    max_overlap = 0

    for syn in synsets:
        definition_words = set(word_tokenize(syn.definition().lower()))
        overlap = len(context.intersection(definition_words))

        if overlap > max_overlap:
            max_overlap = overlap
            best_synset = syn

    return best_synset
```

```python
# Example 1 - bank
sent1 = "I went to the bank to deposit money"
sent2 = "The river overflowed the bank after heavy rain"
```

```python
word = "bank"
```

```python
sense1 = lesk_algorithm(sent1, word)
sense2 = lesk_algorithm(sent2, word)
```

```python
print("Sentence 1:", sent1)
print("Sense 1:", sense1)
print("Definition 1:", sense1.definition())

print("\nSentence 2:", sent2)
print("Sense 2:", sense2)
print("Definition 2:", sense2.definition())
```

```python
# Example 2 - bat
sentence1 = "He hit the ball with a bat during the cricket match."
sentence2 = "A bat was flying in the dark cave at night."
```

```python
word = 'bat'
```

```python
sense3 = lesk_algorithm(sentence1, word)
sense4 = lesk_algorithm(sentence2, word)
```

```python
print("Sentence 1:", sentence1)
print("Sense 1:", sense3)
print("Definition 1:", sense3.definition())

print("\nSentence 2:", sentence2)
print("Sense 2:", sense4)
print("Definition 2:", sense4.definition())
```
